```python
%%R
# Load libraries
install.packages("gridExtra", repos = "http://cran.us.r-project.org")
library(ggplot2)
library(gridExtra)

# Define the function
get_legend <- function(myggplot) {
    tmp <- ggplot_gtable(ggplot_build(myggplot))
    leg <- which(sapply(tmp$grobs, function(x) x$name) == "guide-box")
    legend <- tmp$grobs[[leg]]
    return(legend)
}
```

```python
# Function for making density plots of a single annotation
makeDensityPlot <- function(dataframe, xvar, split, xmin=min(dataframe[xvar], na.rm=TRUE),
xmax=max(dataframe[xvar], na.rm=TRUE), alpha=0.5) {
  # Check if 'split' is missing
  if (missing(split)) {
    return(
      ggplot(data = dataframe, aes_string(x = xvar)) +
        xlim(xmin, xmax) +
        geom_density()
    )
  } else {
    return(
      ggplot(data = dataframe, aes_string(x = xvar, fill = split)) +
        xlim(xmin, xmax) +
        geom_density(alpha = alpha)
    )
  }
}
```

**Lọc dựa trên QUAL và đánh giá kết quả dựa trên mức độ đồng thuận (concordance) với truth set**

```python
# Iterate on thresholds to improve precision while maintaining high sensitivity.
!${TOOLS}/gatk-4.2.0.0/gatk VariantFiltration \
-V ${DATA}/variant-filtering/motherSNP.vcf.gz \
-filter "QD < 2.0" --filter-name "QD2" \
-filter "QUAL < 30.0" --filter-name "QUAL30" \
-filter "SOR > 3.0" --filter-name "SOR3" \
-filter "FS > 60.0" --filter-name "FS60" \
-filter "MQ < 40.0" --filter-name "MQ40" \
-filter "MQRankSum < -12.5" --filter-name "MQRankSum-12.5" \
-filter "ReadPosRankSum < -8.0" --filter-name "ReadPosRankSum-8" \
-O ${DATA}/variant-filtering/motherSNPfilters.vcf.gz
```

```python
!zcat < ${DATA}/variant-filtering/motherSNPfilters.vcf.gz | grep -v '##' | grep -v 'PASS' | head -n20 | cut -f6-10
```

```python
!${TOOLS}/gatk-4.2.0.0/gatk Concordance \
-eval ${DATA}/variant-filtering/motherSNPfilters.vcf.gz \
-truth ${DATA}/resources/motherGIABsnps.vcf.gz \
-L ${DATA}/resources/motherHighconf.bed \
-S ${DATA}/variant-filtering/motherSNPfilters.txt
```
